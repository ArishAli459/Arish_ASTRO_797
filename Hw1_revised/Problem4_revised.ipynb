```python
#Problem 4
import math
import matplotlib.pyplot as plt
import numpy as np

#define variables
c = 299792.458 # km/s

#density parameters
rho_m = 0.3
rho_lam = 0.7
h = 0.7
#Hubble's constant, km/s/Mpc
H_0 = 100*h
#inverse Hubble parameter function
def E_inv(z):
  H =  1/ np.sqrt(rho_m*(1+z)**3 + rho_lam)
  return H

#hubble distance
def D_H(h):
    return 3000./h
    
def D_c(z,h):
    #generate an array of redshifts out to the redshift of interest
    zint = np.linspace(0.0, z, num=100)
    #Hubble distnace

    #comoving distance
    D_C = D_H(h) * np.trapezoid(E_inv(zint),zint)
    return D_C
def D_A(z):
    #compute angular diamter distance at a given redshift
    D_a = D_c(z,h)/(1+z)
    return D_a
def dV_c(z):
    #comoving volume element
    dv= D_H(h) * (1+z)**2 *D_A(z)**2 * E_inv(z)
    return dv
#now make a function for the solid angle
def S_ang(area):
    #area is in square degrees, so we sqrt it to get a side in degrees
    side_deg = np.sqrt(area)
    #now conver to rads
    rad = side_deg * np.pi/180
    #steradians is just rad^2
    str = rad**2
    return str

def V_c(z1,z2,area):
    #define range over redshift for the integral for minimum z value
    zint1 = np.linspace(0,z1, num=100)
    #find the interior of the integral
    integrand1 = dV_c(zint1)
    #carry out the integral
    Vol1 = np.trapezoid(integrand1, x=zint1)
    
    #repeat for max z value
    zint2 = np.linspace(0,z2, num=100)
    integrand2 = dV_c(zint2)
    Vol2 = np.trapezoid(integrand2, x=zint2) 

    #find differnce in volumes
    Voldiff = Vol2-Vol1

    #find comoving volume
    V_c = S_ang(area) * Voldiff
    return V_c

    
```

```python
z = np.array([0.05,1.,2.])
area= 0.25

Vc1 = f"{V_c(z[0], z[1], area):.3e}"
Vc2 = f"{V_c(z[1], z[2],area):.3e}"

print("The comoving volume from 0.05<z<1 is " +Vc1)
print("The comoving volume from 1<z<2 is " +Vc2)
        
```
