```python
#from the textbook, the number of stars can be found by integrating dN = phi(m)dm, where phi(m) is the IMF and dm is the mass range

#from the Chabrier paper, lets define a function that calculates and integrates the IMF

"""
def chab(m):
    #there are two ranges for the IMF, m<= m_norm, or m>m_norm, where m_norm=1 M_sun. Lets go the m<=m_norm first
    #for this one the IMF is Aexp [-(log m - log m_c)**2 ) /2sigma**2] 
    if m <=1.0:
        A = .158
        mc = 0.079
        sig = 0.69
        e = np.exp(-(np.log10(m)-np.log10(mc))**2/(2*sig**2))
        N = A * e
        return N
    else:
        #for this range of m>1, the IMF is Am**(-x)
        A= 4.4 * 10**(-2)
        x = 1.3
        N = A*m**(-x)
        return N

        """
def chab(m):
    #there are two ranges for the IMF, m<= m_norm, or m>m_norm, where m_norm=1 M_sun. Lets go the m<=m_norm first
    #for this one the IMF is Aexp [-(log m - log m_c)**2 ) /2sigma**2] 
    if m <=1.0:
        A = .158
        mc = 0.079
        sig = 0.69
        e = np.exp(-(m-mc)**2/(2*sig**2))
        N = A * e
        return N
    else:
        #for this range of m>1, the IMF is Am**(-x)
        A= 4.4 * 10**(-2)
        x = 1.3
        N = A*m**(-x)
        return N       
        
# Integrands: Number, Mass, Luminosity 
n_func = lambda m: chab(m)
m_func = lambda m: m * chab(m)
#the mass-luminosity relation for different ranges of mass multiplied by the IMF
def l_func(m):
    if m < 0.43:
        return 0.23 * (m**2.3) * chab(m)
    elif m < 2.0:
        return (m**4) * chab(m)
    else:
        return 1.4 * (m**3.5) * chab(m)

# Create arrays for numerical integration
m_values1 = np.logspace(np.log10(0.1), np.log10(1.0), 2000)  # Create array of mass values for range 1
m_values2 = np.logspace(np.log10(1.0), np.log10(10.0), 2000)  # Create array of mass values for range 2

# Calculate function values for each mass point
n_values1 = np.array([n_func(m) for m in m_values1])
n_values2 = np.array([n_func(m) for m in m_values2])
m_values_n1 = np.array([m_func(m) for m in m_values1])
m_values_n2 = np.array([m_func(m) for m in m_values2])
l_values1 = np.array([l_func(m) for m in m_values1])
l_values2 = np.array([l_func(m) for m in m_values2])

# Computing Integrals correctly with arrays
N1 = np.trapezoid(n_values1, m_values1)  
N2 = np.trapezoid(n_values2, m_values2)
M1 = np.trapezoid(m_values_n1, m_values1)
M2 = np.trapezoid(m_values_n2, m_values2)
L1 = np.trapezoid(l_values1, m_values1)
L2 = np.trapezoid(l_values2, m_values2)

print(f"Number Ratio: {N2/N1:.4f}")
print(f"Mass Ratio: {M2/M1:.4f}")
print(f"Luminosity Ratio: {L2/L1:.4f}")      
```
